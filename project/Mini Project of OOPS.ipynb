```python
class character:

    def __init__(self,name):
        self.name=name
        self.__life=3
        self.__score=0
    def kicked (self):
        self.__score+=10
    def punched (self):
        self.__score+=5
    def stabbed (self):
        self.__life-=1
    def displaylife(self):
        return f" The {self.name} has {self.__life} life "
    def displayscore(self):
        return f" The {self.name} has {self.__score} score"

obj = character("Mario")
(obj.kicked())
(obj.stabbed())
(obj.punched())
(obj.stabbed())
print(obj.displaylife())
print(obj.displayscore())
```
